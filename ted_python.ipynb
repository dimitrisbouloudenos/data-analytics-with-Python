```python
###visualise of distribution of comments
df[df['comments']>1000]['comments'].plot(kind='bar')
```

```python
###plot of number of talks took place in year
df['film_datetime']= pd.to_datetime(df['film_date'],unit='s')
df['year']=df['film_datetime'].dt.year
```

```python
df['year'].value_counts().plot(kind='bar')
```

```python
####what were is best events in tend history
df.groupby('event')['views'].agg(['count','mean','sum']).sort_values('sum').tail()
```

```python
###What is the average number of views per speaker?
df.groupby('main_speaker')['views'].mean().sort_values(ascending=False).head(10)
```

```python
###Which speakers have given the most talks? (Top 10)
df['main_speaker'].value_counts().head(10).plot(kind='barh',title='top 10 Speakers by Number of Talks')

```

```python
###What is the average number of views per year? (We need to extract the year from the film_date
df.groupby('year')['views'].mean().plot(kind='line',title='Average Views per Year')
```

```python
###Top 10 most viewed TED Talks
df[['name','main_speaker','views']].sort_values(by='views',ascending=False).head(10)
```

```python
###Correlation between views and comments
df.plot.scatter(x='comments',y='views',alpha=0.5)
```

```python
###Duration of talks:is there a pattern
df['duration_minutes']=df['duration']/60
df['duration_minutes'].plot(kind='hist',bins=30,title='Talk Duration(minutes)')
```
