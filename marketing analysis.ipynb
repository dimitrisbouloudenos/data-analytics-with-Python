```python
# What is the distribution of the target variable ('y')? Is the dataset balanced or imbalanced?
df['y'].value_counts(normalize=True)
```

```python
# Visualize the target variable distribution
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x='y',data=df)
plt.title('Target Varieble')
plt.show()
```

```python
# Does the contact method (e.g. cellular, telephone) affect the response?
contact_response = df.groupby('contact')['y'].value_counts(normalize=True).unstack().fillna(0)
contact_response['yes'].sort_values(ascending=False).plot(kind='bar')
plt.title('Response Rate by Contact Method')
plt.ylabel('Proportion of "yes"')
plt.show()
```

```python
# What is the response rate ('yes') for each job category?
job_response = df.groupby('job')['y'].value_counts(normalize=True).unstack().fillna(0)
job_response['yes'].sort_values(ascending=False).plot(kind='bar', figsize=(10,5))
plt.title('Response Rate by Job')
plt.ylabel('Proportion of "yes"')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Select only numeric columns
numeric_df=df.select_dtypes(include=['int64','float64'])
# Create the correlation matrix
corr_matrix=numeric_df.corr()
# Plot the heatmap
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm',fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
