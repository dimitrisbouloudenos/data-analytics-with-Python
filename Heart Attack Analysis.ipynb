```python
## Function to create a new column to find risk of Blood Sugar
def Sugar(x):
    if x>200:
        return 'High'
    else:
        return 'Low'
```

```python
df['Sugar_Risk']=df['Blood sugar'].apply(Sugar)
```

```python
### Categorize Heart Rate to low,normal,high
def Heart_Rate(x):
    if x>100:
        return 'High'
    elif x>60 and x<100:
        return 'Medium'
    else:
        return 'Low'
```

```python
df['Categorize_Heart_Rate']=df['Heart rate'].apply(Heart_Rate)
```

```python
df['Categorize_Heart_Rate'].value_counts()
```

```python
###Visualize Result
df['Categorize_Heart_Rate'].value_counts().plot(kind='bar')
```
