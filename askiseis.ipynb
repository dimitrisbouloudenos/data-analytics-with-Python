```python
import matplotlib.pyplot as plt
```

```python
monthly_crimes=df.groupby('month')['no_of_crimes'].mean()
```

```python
plt.figure(figsize=(10,5))
monthly_crimes.plot(kind='bar',color='salmon',edgecolor='black')
plt.title('mesos arithmos eglimaton ana mina')
plt.ylabel('mo eglimaton')
plt.xticks(rotation=0)
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
london_data=df[df['area']=='London']
london_grouped=london_data.groupby('date')['average_price'].mean()
plt.figure(figsize=(12,5))
london_grouped.plot(color='teal')
plt.title('mesi timi spitiou sto london')
plt.xlabel('imerominia')
plt.ylabel('mesi timi')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Φιλτράρισμα για London
london_data = df[df['area'] == 'London']

# Group by date & μέση τιμή
london_grouped = london_data.groupby('date')['average_price'].mean()

# Plot
plt.figure(figsize=(12, 5))
london_grouped.plot(color='teal')
plt.title('Μέση Τιμή Σπιτιού στο London με την Πάροδο του Χρόνου')
plt.xlabel('Ημερομηνία')
plt.ylabel('Μέση Τιμή (£)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
